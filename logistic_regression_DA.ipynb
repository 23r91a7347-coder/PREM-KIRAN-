```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression


data = {
    'Age': [25, 30, 22, 40, 35],
    'Salary': [50000, 54000, 58000, 61000, 59000],
}

df = pd.DataFrame(data)
print("Original Data:\n", df)


df['HighSalary'] = (df['Salary'] >= 58000).astype(int)
X = df[['Age']]
y = df['HighSalary']

log_model = LogisticRegression()
log_model.fit(X, y)

print("\nModel Coefficient:", log_model.coef_[0][0])
print("Model Intercept:", log_model.intercept_[0])


age_range = np.linspace(df['Age'].min(), df['Age'].max(), 200)
age_df = pd.DataFrame(age_range, columns=['Age'])

probabilities = log_model.predict_proba(age_df)[:, 1]


plt.scatter(df['Age'], df['HighSalary'], color='blue', label="Actual Classes")
plt.plot(age_range, probabilities, color='red', label="Logistic Curve")

plt.xlabel("Age")
plt.ylabel("Probability of High Salary")
plt.title("Logistic Regression: Age vs High Salary Probability")
plt.grid(True)
plt.legend()
plt.show()
```
